```python
# Collect user movie preferences
users = []

n = int(input("Enter number of users: "))

for i in range(n):
    print(f"\nUser {i+1}")
    name = input("Enter user name: ").strip()
    movies = input("Enter favorite movies (comma-separated): ").split(",")

    # Clean movie names and store as tuple
    movies = tuple(movie.strip() for movie in movies)
    users.append((name, movies))
```

```python
#Users who share at least two movies in common
print("Users who share at least two movies in common:")

found = False
for i in range(len(users)):
    for j in range(i + 1, len(users)):
        user1, movies1 = users[i]
        user2, movies2 = users[j]

        common_movies = set(movies1).intersection(set(movies2))

        if len(common_movies) >= 2:
            print(f"{user1} and {user2} (Common movies: {common_movies})")
            found = True

if not found:
    print("No users share at least two movies.")
```

```python
#Unique movies
all_movies = set()

for _, movies in users:
    all_movies.update(movies)

print("All Unique Movies in the Database:")
for movie in all_movies:
    print(movie)
```

```python
#Movies favourite by only one user
movie_count = {}

for _, movies in users:
    for movie in movies:
        movie_count[movie] = movie_count.get(movie, 0) + 1

unique_to_one_user = [
    movie for movie, count in movie_count.items() if count == 1
]

print("\nMovies marked as favorite by only one user:")
if unique_to_one_user:
    for movie in unique_to_one_user:
        print(movie)
else:
    print("None")
```
