```python
# Sports Event Participation Summary

participants = []

n = int(input("Enter number of participants: "))

# Step 1: Collect participant data
for i in range(n):
    print(f"\nParticipant {i+1}")
    name = input("Enter participant name: ").strip()
    sports = input("Enter sports participated in (comma-separated): ").split(",")

    # Clean sports names
    sports = [sport.strip() for sport in sports]

    # Store as tuple
    participants.append((name, sports))
```

```python
print("Participants playing both Cricket and Football:")
found = False

for name, sports in participants:
    if "cricket" in sports and "football" in sports:
        print(name)
        found = True

if not found:
    print("None")
```

```python
all_sports = set()

for _, sports in participants:
    all_sports.update(sports)

print("Total number of unique sports being played:", len(all_sports))
print("Unique Sports:", all_sports)
```

```python
sport_count = {}

for _, sports in participants:
    for sport in sports:
        sport_count[sport] = sport_count.get(sport, 0) + 1

max_count = max(sport_count.values())

most_played_sports = tuple(
    sport for sport, count in sport_count.items() if count == max_count
)

print("Sport(s) with the highest participation count:")
print(most_played_sports)
```
