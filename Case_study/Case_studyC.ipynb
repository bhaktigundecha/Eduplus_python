```python
# Initialize storage
registration_data = {}

# User enters data
num_students = int(input("Enter number of students: "))
for _ in range(num_students):
    name = input("\nEnter Student Name: ")
    # Input courses as a comma-separated string and convert to a tuple
    courses_input = input(f"Enter courses for {name} (e.g. AI, Data Science): ")
    registration_data[name] = tuple(c.strip().title() for c in courses_input.split(","))

```

```python
# 1. Identify all students enrolled in both ‘Data Science’ and ‘AI’
target_courses = {"Data Science", "AI"}
print("\n--- Students enrolled in both Data Science and AI ---")
for student, courses in registration_data.items():
    # Convert tuple to set for comparison
    if target_courses.issubset(set(courses)):
        print(f"- {student}")
```

```python
# 2. Find the total number of unique courses offered
all_courses = []
for courses in registration_data.values():
    all_courses.extend(courses)

unique_courses = set(all_courses)
print(f"\nTotal Unique Courses Offered: {len(unique_courses)}")
print(f"Course List: {', '.join(sorted(unique_courses))}")
```

```python
# 3. Find all courses that no student has repeated
# Logic: Find courses that appear only once across the entire university
from collections import Counter
course_counts = Counter(all_courses)

unique_enrollments = [course for course, count in course_counts.items() if count == 1]
print("\n--- Courses with no student repeats (Only one enrollment) ---")
if unique_enrollments:
    print(", ".join(unique_enrollments))
else:
    print("No such courses; all courses have at least two students.")
```
