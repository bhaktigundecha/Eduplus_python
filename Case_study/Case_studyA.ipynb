```python
#Case Study A: Employee Skill Tracker (HR Data Filtering)
```

```python
#python program for input the employee details
employees = {}
num_employees = int(input("Enter the number of employees: "))
for i in range(num_employees):
    print(f"\n--- Entering details for Employee #{i + 1} ---")
    name = input("Enter Employee Name: ")
    skills_input = input("Enter skills (separated by commas, e.g., Python, SQL, Java): ")
    skills_list = [skill.strip() for skill in skills_input.split(",")]
    employees[name] = skills_list
print("\n--- Employee Skills Registry ---")
for name, skills in employees.items():
    print(f"Employee: {name} | Skills: {', '.join(skills)}")    
```

```python
#1.Display all employees who know both 'Python' and 'SQL'
```

```python
target_skills = {"Python", "SQL"}
print("\n--- Employees knowing both Python and SQL ---")
found_any = False
for name, skills in employees.items():
    if target_skills.issubset(skills):
        print(f" {name}")
        found_any = True
if not found_any:
    print("No employees found with both skills.")
```

```python
# Find and print the unique skills available across all employees
```

```python
unique_skills = set()
for skills in employees.values():
    unique_skills.update(skills)

print("\n--- All Unique Skills in the Organization ---")
print(", ".join(sorted(unique_skills)))
```

```python
# employees that shares at least one  skill with a specific employee entered by the user
```

```python
search_name = input("\nEnter an employee name to check for shared skills: ")

if search_name in employees:
    # 1. Convert the base employee's skills to a set once
    base_skills = set(employees[search_name])
    count = 0
    print(f"Employees sharing at least one skill with {search_name}:")
    
    for name, skills_list in employees.items():
        if name != search_name:
            # 2. Convert current employee's skill list to a set
            current_skills = set(skills_list)
            
            # 3. Use isdisjoint() on the set object
            if not base_skills.isdisjoint(current_skills):
                # 4. Use & operator on sets to find common skills
                shared = base_skills & current_skills
                print(f"- {name} (Shared: {shared})")
                count += 1
    
    print(f"\nTotal employees sharing skills with {search_name}: {count}")
else:
    print(f"Employee '{search_name}' not found in the registry.")
```
