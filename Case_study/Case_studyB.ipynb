```python
#Case Study B: Product Inventory Analyzer (Warehouse Comparison)
```

```python
def get_inventory(warehouse_name):
    inventory = []
    print(f"\nEnter details for {warehouse_name}:")
    n = int(input("How many products? "))
    for _ in range(n):
        p_id = input("Product ID: ")
        p_name = input("Product Name: ").strip().lower()
        qty = int(input("Quantity: "))
        inventory.append({"ID": p_id, "Name": p_name, "Quantity": qty})
    return inventory
```

```python
# 1. Accept product details
w1 = get_inventory("Warehouse 1")
w2 = get_inventory("Warehouse 2")

# Create sets of names for comparison
names1 = {p['Name'] for p in w1}
names2 = {p['Name'] for p in w2}
```

```python
# 2. Display common products (Intersection)
common_names = names1.intersection(names2)
print("Common Products")
for name in common_names:
    print(name.capitalize())
```

```python
# 3. Display unique products across both (Symmetric Difference)
unique_names = names1.symmetric_difference(names2)
print("Unique Products (Not in both)")
for name in unique_names:
    print(name.capitalize())
```

```python
#4. Combine and sort by quantity descending
combined = w1 + w2
# Using sorted() with reverse=True for descending order
sorted_combined = sorted(combined, key=lambda x: x['Quantity'], reverse=True)

print("Combined Inventory (Sorted by Quantity Descending)")
print(f"{'ID':<10} {'Name':<10} {'Quantity':<10}")
for p in sorted_combined:
    print(f"{p['ID']:<10} {p['Name'].capitalize():<10} {p['Quantity']:<10}")
```
