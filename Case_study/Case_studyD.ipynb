```python
#Case Study D: Customer Purchase Insights (E-commerce Scenario)
```

```python
#Collect customer purchase data
purchases = []

n = int(input("Enter number of customers: "))

for i in range(n):
    print(f"\nCustomer {i+1}")
    name = input("Enter customer name: ").strip()
    products = input("Enter products purchased (comma-separated): ").split(",")

    # Clean product names
    products = [p.strip() for p in products]

    # Store data as tuple
    purchases.append((name, products))
```

```python
#Total number of unique product brought
all_products = set()

for _, product_list in purchases:
    all_products.update(product_list)

print("Total number of unique products bought:", len(all_products))
print("Unique Products:", all_products)
```

```python
#Customer who purchased both laptop and headphones
print("Customers who purchased both Laptop and Headphones:")
found = False

for customer, product_list in purchases:
    if "Laptop" in product_list and "Headphone" in product_list:
        print(customer)
        found = True

if not found:
    print("None")
```

```python
#3. Most frequently purchased product(s)
product_frequency = {}
for _, product_list in purchases:
    for product in product_list:
        product_frequency[product] = product_frequency.get(product, 0) + 1

max_count = max(product_frequency.values())

most_frequent_products = [
    product for product, count in product_frequency.items()
    if count == max_count
]

print("Most Frequently Purchased Product(s):")
for product in most_frequent_products:
    print(f"{product} (purchased {max_count} times)")
```
