```python
#Case Study E: Travel Planner (Data Deduplication Challenge)
```

```python
# 1: Collect booking details
bookings = set()   # Using a set to automatically remove duplicates
destination_list = []  # To track destination frequency

n = int(input("Enter the number of bookings: "))

for i in range(n):
    print(f"\nBooking {i+1}")
    traveler = input("Enter Traveler Name: ").strip()
    destination = input("Enter Destination: ").strip()
    date = input("Enter Date (YYYY-MM-DD): ").strip()

    # Add booking as a tuple (immutable, suitable for set)
    bookings.add((traveler, destination, date))
    destination_list.append(destination)
```

```python
# 2: Display deduplicated booking records
print("Unique Booking Records:")
for booking in bookings:
    print(booking)
```

```python
#3: Display all unique destinations
unique_destinations = set(destination_list)
print("Unique Destinations:")
for dest in unique_destinations:
    print(dest)
```

```python
#4: Find top 3 most visited destinations
destination_count = {}

for dest in destination_list:
    destination_count[dest] = destination_count.get(dest, 0) + 1

# Sort destinations by frequency (descending)
top_3 = sorted(destination_count.items(), key=lambda x: x[1], reverse=True)[:3]

print("\nTop 3 Most Visited Destinations:")
for dest, count in top_3:
    print(f"{dest}: {count} visits")
```
