```python
# Student Attendance Analyzer

attendance = []  # List to store attendance sets for each day
days = 5
```

```python
# 1: Input attendance for 5 days
for day in range(1, days + 1):
    print(f"\nEnter students present on Day {day} (comma-separated):")
    students = input().split(",")

    # Clean and store as a set
    day_set = {student.strip() for student in students}
    attendance.append(day_set)
```

```python
#student present everyday
present_every_day = set.intersection(*attendance)

print("Students present every day:")
if present_every_day:
    for student in present_every_day:
        print(student)
else:
    print("None")
```

```python
#student absent on atleast one day
all_students = set.union(*attendance)
absent_at_least_one_day = all_students - present_every_day

print("\nStudents absent on at least one day:")
for student in absent_at_least_one_day:
    print(student)
```

```python
#Total unique student across 5 days
print("Total unique students across all five days:", len(all_students))
print("Unique Students:", all_students)
```
