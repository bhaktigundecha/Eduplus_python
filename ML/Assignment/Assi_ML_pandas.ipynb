```python
# 2. Create a DataFrame with columns: "Student_Name", "Maths", "Science", "English".
#     Fill it with 5 sample student records and display the top 3 rows using head().
data = {
    "Student_Name": ["Bhakti", "Isha", "Rutuja", "Saniya", "Aditi"],
    "Maths_score": [99,91,90,89,78],
    "Science_score": [90,92,89,93,67],
    "English_score": [98,88,90,85,78]
}
df = pd.DataFrame(data)
print("The data frame is:\n", df)
print("\nTop 3 rows are:\n", df.head(3))
```

```python
#3.Import a CSV file (use any sample dataset) into a DataFrame.
#Display the number of rows and columns using shape, and print column names
print("Number of rows and columns in data frame is:", df.shape)
print("Columns in data frame are:", df.columns)
```

```python
# 4. Add a new column named "Total_Marks" that sums all three subject scores.
#     Then create another column "Average" by dividing Total_Marks by 3.
```

```python
df["Total_marks"] = df["Maths_score"]+df["English_score"]+df["Science_score"]
df["Average"] = df["Total_marks"]/3
print("Updated data frame is:\n", df)
```

```python
# 5. Introduce a missing value (NaN) in the "Science" column and then:
#     (a) Identify missing values using isnull()
#     (b) Replace the missing value with the column mean using fillna().
```

```python
df.loc[2, "Science_score"] = None
print("\nData frame with missing value:\n", df)
print(df.isnull()) #a
df["Science_score"].fillna(df["Science_score"].mean(), inplace=True) #b
print("\ndata frame after using fillna():\n",df)
```

```python
# 6. Filter and display all students whose "Average" score is greater than 75.
```

```python
print("Student whose average is greater than 75:\n", df[df["Average"] > 75])
```

```python
# 7. Sort the DataFrame by "Total_Marks" in descending order and show the top 3 performers.
```

```python
sorted_df = df.sort_values(by = "Total_marks", ascending=False)
print("The top 3 performers are:\n", sorted_df.head(3))
```

```python
#8. Group students based on their "English" score range (like above or below 70)
#     and find the mean of "Maths" and "Science" marks for each group.
```

```python
df['English_Range'] = df['English_score'].apply(lambda x: 'Above 70' if x >= 70 else 'Below 70')
result = df.groupby('English_Range')[['Maths_score', 'Science_score']].mean()
print(result)
```

```python
#9. Merge two DataFrames:
#     (a) One with "Student_Name" and "Total_Marks"
#     (b) Another with "Student_Name" and "Sports_Score"
#     Then create a new column "Overall_Score" = Total_Marks + Sports_Score.
```

```python
data_dict = pd.DataFrame({
    "Student_Name": ["Bhakti", "Isha", "Rutuja", "Saniya", "Aditi"],
    "Sport_score": [70,80,75,80,80]
})
merged = pd.merge(df, data_dict, on="Student_Name")
print(merged)
```

```python
merged["Overall_score"] = merged["Total_marks"]+merged["Sport_score"]
print(merged)
```

```python
#10. Export the final DataFrame to a CSV file on your Desktop (use os library to define path)
#     and verify if the file is successfully created using os.path.exists().
```

```python
import os

desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")
file_name = "Final_Student_Data.csv"
full_path = os.path.join(desktop_path, file_name)

merged.to_csv(full_path, index=False)

if os.path.exists(full_path):
    print(f"Success! The file has been created at: {full_path}")
else:
    print("Error: The file was not created.")
```
