```python
#Case Study: Employee Record Management Using Dictionary(ISHA KADAM)
#Write a Python program using a dictionary to store employee records. Each employee record should contain the employee name, department, and salary.
```

```python
#Employee data
employee_data = {
    "101": {"name": "Isha", "department": "IT", "salary": 60000},
    "102": {"name": "Bhakti", "department": "HR", "salary": 45000},
    "103": {"name": "Saniya", "department": "Finance", "salary": 55000},
    "104": {"name": "Rutuja", "department": "IT", "salary": 70000},
    "105": {"name": "Anita", "department": "HR", "salary": 50000}
}
```

```python
#Display all employees
print("All employee details :", employee_data)
```

```python
# Find total numbers of employees
print("Total Employees are:", len(employee_data))
```

```python
#Display names of all employees
print("Employee Names are:")
for emp in employee_data.values():
    print(emp["name"])
```

```python
#Find employee working in IT Department
print("Employee in IT department are:")
for emp in employee_data.values():
    if emp["department"] == "IT":
        print(emp["name"])
```

```python
#Calculate total salary expences
total_salary = 0
for emp in employee_data.values():
    total_salary+=emp["salary"]
print("Total salary expenses:", total_salary)    
```

```python
#Find highest paid employee
highest_paid = max(employee_data.values(), key=lambda x:x["salary"])
print("Highest paid employee is:", highest_paid["name"],highest_paid["salary"])
```

```python
#Find lowest paid employee
lowest_paid = min(employee_data.values(), key=lambda x:x["salary"])
print("Lowest paid employee is:", lowest_paid["name"],lowest_paid["salary"])
```

```python
#Increase salary of IT employee by 10%
for emp in employee_data.values():
    if emp["department"] == "IT":
        emp["salary"]+=emp["salary"]*0.10
        print(emp["name"],emp["department"], emp["salary"])
```

```python
#Count employee department wise
dept_count = {}
for emp in employee_data.values():
    dept = emp["department"]
    dept_count[dept] = dept_count.get(dept,0)+1
print("Department wise Employee count is:", dept_count)    
```

```python
#Add new employee 
employee_data["106"] = {
    "name":"Suresh",
    "department":"Finance",
    "salary":48000
}
print("Updated employee data:")
print(employee_data)
```

```python
#Case study:Student Performance Management System(BHAKTI GUNDECHA)
#A university wants to build a Student Performance Management System to track 
#student's academic records across multiple subjects.
#Each student has an unique student ID, is enrolled in multiple subjects, has marks, attendance, and grades
#Can update and drop subjects, needs performance analytics
```

```python
#Student dictionary
students = {
    "101": {
        "name": "Amit",
        "subjects": {
            "Math": {"marks": 85, "attendance": 90},
            "Physics": {"marks": 78, "attendance": 88}}},
    "102": {
        "name": "Neha",
        "subjects": {
            "Math": {"marks": 92, "attendance": 95},
            "Chemistry": {"marks": 81, "attendance": 85}}}}
print("Student details are:")   
print(students)
```

```python
#1.Add a new student with ID, name, and an empty subject list
students["103"] = {
    "name": "Rahul",
    "subjects": {"Biology":{"marks":78, "attendance":74}}}
print("New Student:",students["103"])
print("All students are:", students)
```

```python
#2. Enroll Student in a New Subject
students["103"]["subjects"]["Chemistry"] = {
    "marks": 88,
    "attendance": 92
}
print("After Enrolling Subject:")
print(students["103"])
```

```python
#3. Update Marks for a Subject
students["101"]["subjects"]["Math"]["marks"] = 90
print("After Updating Marks:")
print(students["101"]["subjects"]["Math"])
```

```python
#4. Safely Retrieve Student Details
student = students.get("104", "Student Not Found")
print("Student Search Result:")
print(student)
```

```python
#5. Remove a Subject
removed_subject = students["102"]["subjects"].pop("Chemistry")
print("Removed Subject Details:")
print(removed_subject)
```

```python
#6. Delete a Student Record
students.pop("103")
print("After Removing Student S103:")
print(students.keys())
```

```python
#7. Calculate Average Marks per Student
print("Average Marks:")
for sid, info in students.items():
    total = 0
    count = 0
    for sub in info["subjects"].values():
        total += sub["marks"]
        count += 1
    avg = total / count
    print(sid, ":", avg)
```

```python
#8. Find Subject Topper
subject = "Math"
topper = None
highest = 0
for sid, info in students.items():
    if subject in info["subjects"]:
        marks = info["subjects"][subject]["marks"]
        if marks > highest:
            highest = marks
            topper = info["name"]
print("Math Topper:")
print(topper, "-", highest)
```

```python
#9. Generate Grade Report
grades = {}
for sid, info in students.items():
    total = sum(sub["marks"] for sub in info["subjects"].values())
    avg = total / len(info["subjects"])
    if avg >= 90:
        grades[sid] = "A"
    elif avg >= 80:
        grades[sid] = "B"
    else:
        grades[sid] = "C"
print("Grade Report:")
print(grades)
```

```python
#10. Subject-wise Student Count
subject_count = {}
for info in students.values():
    for subject in info["subjects"]:
        subject_count.setdefault(subject, 0)
        subject_count[subject] += 1
print("Subject-wise Student Count:")
print(subject_count)
```
