```python
student_details = {
    "Isha":{"age":22, "grade":90, "department":"IT"},
    "Bhakti":{"age":21, "grade":98, "department":"AIDS"},
    "Rutuja":{"age":20, "grade":91, "department":"IT"}
} #23
print("The nested dictionary is:", student_details)
```

```python
print("The grade of Isha is:", student_details["Isha"]["grade"]) #24
```

```python
student_details["Rutuja"]["department"] = "CS" #25
print("The updated dict is:", student_details)
```

```python
print("Student names and grades:") #26
for name,details in student_details.items():
    print(f"Name:{name} Grade:{details["grade"]}")
```

```python
new_dict = {**student, **student_details} #27
print("Two dict after merge is:", new_dict)
```

```python
new_dict.clear() #28
print(new_dict)
```
