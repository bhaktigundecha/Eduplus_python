```python
def global_var(): #2
    global subtotal, tax, discount, total, payment_method

    subtotal = 2420
    tax = 121.0
    discount = 242.0
    payment_method = "UPI"
    total = subtotal + tax - discount

global_var()

print(f"Subtotal: {subtotal}")
print(f"Tax: {tax}")
print(f"Discount: {discount}")
print(f"Total: {total}")
print(f"Payment: {payment_method}")
```

```python
def counter(): #6
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment()
print(counter())
```

```python
def tracker(): #13
    total = 0
    def add(n):
        nonlocal total
        total += n
        print("Running Total:", total)
    return add

t = tracker()
t(10)
t(20)
```

```python
#24
def call_tracker():
    count = 0
    def wrapper():
        nonlocal count
        count += 1
        print(f"Call count: {count}")
    return wrapper

track = call_tracker()
track()
track()
track()
```

```python
#26
def expense_manager():
    total_exp = 0
    def add_exp(amount):
        nonlocal total_exp
        total_exp += amount
        print(f"Total Expenses: {total_exp}")
    return add_exp

manager = expense_manager()
manager(50)
manager(150)
```
