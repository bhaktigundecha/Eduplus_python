```python
#1
num = 1
while num<=10:
    print(num)
    num+=1
```

```python
#2
num = 1
while num<=50:
    if num % 2 ==0:
        print(num)
    num+=1
```

```python
#3
num = 1
sumN = 0
while num<=20:
    sumN+=num
    num+=1
print("Sum of first 20 natural numbers are:", sumN)
```

```python
#4
num = 1
while num<=10:
    print("7 * ", num, " = ", 7*num)
    num+=1
```

```python
#5
num = 123456
reversed_num = 0

while num != 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10 

print("Reversed Number:", reversed_num)
```

```python
#6
num = 12345678
count = 0
while num != 0:
    num //=10
    count +=1
print(count)
```

```python
#7
num = 5
factorial = 1
i = 1
while num>=i:
    factorial*=i
    i+=1
print(factorial)
```

```python
#8
num = 50
a, b = 0, 1
while a<=num :
    print(a)
    a, b = b, a + b  # Simultaneous update
```

```python
#9
num = 12321
original = num
reversed_num = 0

while num != 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10
print(reversed_num)
if reversed_num == original:
    print("is palindrome")
else:
    print("Not palindrome")
```
